```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 1. titanic 데이터셋(sns.load_dataset('titanic')) 에서
# ['age','sex', 'class', 'fare', 'survived'] 5개 열을 선택해서 df를 만드세요.

df = titanic[['age', 'sex', 'class', 'fare', 'survived']]
df.head()
```

```python
# 2. class 열을 기준으로 분할하고 각 그룹에 대한 모든 열의 표준편차를 집계하여 df2로 반환하세요.

def score(x) :
    return (x - x.mean())

df2 = df.groupby('class').agg({'age' : 'std', 'fare' : 'std', 'survived' : 'std'}).reset_index()
df2
```

```python
# 3. 최대값과 최소값의 차를 계산하는 사용자 정의 함수(minmax) 를 만들고
#    각 그룹의 최대값과 최소값의 차를 계산하여 그룹별로 집계해 보세요.

def minmax(a) :
    result = max(a) - min(a)
    return result

df3 = df.groupby('class').agg(minmax)
df3
```

```python
# 4. 각 열에 2개 함수(min, max)함수를 일괄 적용해 보세요.

df4 = df.groupby('class').agg([min, max])
df4
```

```python
# 5. fare 열에는 min, max 함수를, age 열에는 mean 함수를 적용해 보세요.
 
df.groupby('class').agg({'fare':['min', 'max'], 'age' : 'min'})
```

```python
# 6. age 열의 평균이 30보다 작은 그룹만을 필터링하여 데이터프레임으로 반환하세요, (편집됨) 

a = df.groupby('class').filter(lambda x:x['age'].mean() <= 30)
a.tail(5)
```
